```python
#change to lower case
df['texts']=df['texts'].str.lower()
df
```

```python
#delete number in texts

import re # impor modul regular expression
#use regexr
hasil=[]
for x in df['texts']:
  x =re.sub(r"\d+", "", x)
  hasil.append(x)
# print(hasil)
df['texts']=hasil
pd.set_option("display.max_colwidth", -1)
df
```

```python
#delete link and @
hasil=[]
for x in df['texts']:
  x =re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", x)
  hasil.append(x)
# print(hasil)
df['texts']=hasil
pd.set_option("display.max_colwidth", -1)
df
```

```python
#delete tanda baca
import string

hasil=[]
for x in df['texts']:
  x =x.translate(str.maketrans("","",string.punctuation))
  hasil.append(x)
df['texts']=hasil
pd.set_option("display.max_colwidth", -1)
df
#I still keep # (hastags)
```

#Tokenizing

```python
#create token
def cleaning_tokenizing(txt):
  txt = txt.lower() # menjadi huruf kecil semua
  txt = txt.translate(str.maketrans("","",string.punctuation)) #hapus symbol
  txt = nltk.tokenize.word_tokenize(txt) # tokenize
  return txt
df["texts"].apply(cleaning_tokenizing)
```

#Vectorization (Bag of Words and TF-IDF)

```python
#do CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```
